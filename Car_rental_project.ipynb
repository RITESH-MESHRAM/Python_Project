```python
class CarRentalSystem:
    def __init__(self):
        self.available_cars = {"Toyota": 60, "Renault": 40, "Maruti Suzuki": 70, "Hyundai Creta": 60, "Hyundai Verna": 80}
        self.rental_prices = {"Toyota": 1000, "Renault": 2100, "Maruti Suzuki": 1200, "Hyundai Creta": 1200, "Hyundai Verna": 1400}
        self.rented_cars = {}
        self.receipt = {}

    def show_available_cars(self):
        print("================== Available Cars:===================")
        for car, quantity in self.available_cars.items():
            print(f"{car}: {quantity} Qty.")   

    def show_rental_prices(self):
        print("****************************************\n Rental Prices:")
        for car, price in self.rental_prices.items():
            print(f"{car}: {price} day")

    def rent_car(self, name, mobile, car, quantity,days):
        if car not in self.available_cars:
            print(f"Sorry, {car} is not available for rent.")
        elif quantity > self.available_cars[car]:
            print(f"Sorry, we only have {self.available_cars[car]} {car} available for rent.")
        else:
            self.available_cars[car] -= quantity
            self.rented_cars[car] = quantity
            self.receipt["Name"] = name
            self.receipt["Mobile No."] = mobile
            self.receipt["Rent Car Quantity"] = quantity
            self.receipt["Rent Car Price"] = self.rental_prices[car] * quantity
            print(f"{quantity} {car} rented successfully.")

    def return_car(self, car, quantity):
        if car not in self.rented_cars:
            print(f"Sorry, you have not rented {car}.")
        elif quantity > self.rented_cars[car]:
            print(f"Sorry, you have only rented {self.rented_cars[car]} {car}.")
        else:
            self.available_cars[car] += quantity
            self.rented_cars[car] -= quantity
            print(f"{quantity} {car} returned successfully.")

    def generate_receipt(self):
        print("******************** Receipt: **********************")
        for key, value in self.receipt.items():
            print(f"{key}: {value}")

car_rental_system = CarRentalSystem()

while True:
    print("********************************************************* \nMenu Driven:")
    print("1. Show available cars")
    print("2. Show rental prices")
    print("3. Rent a car")
    print("4. Return a car")
    print("5. Generate receipt")
    print("6. Exit")

    choice = int(input("Select an option: "))

    if choice == 1:
        car_rental_system.show_available_cars()
    elif choice == 2:
        car_rental_system.show_rental_prices()
    elif choice == 3:
        name = input("Enter your name: ")
        mobile = input("Enter your mobile number: ")
        car = input("Enter the car you want to rent: ")
        quantity = int(input("Enter the quantity you want to rent: "))
        days = int(input("Enter the number of days you want to car"))
        car_rental_system.rent_car(name, mobile, car, quantity,days)
    elif choice == 4:
        car = input("Enter the car you want to return: ")
        quantity = int(input("Enter the quantity you want to return: "))
        car_rental_system.return_car(car, quantity)
    elif choice == 5:
        car_rental_system.generate_receipt()
    elif choice == 6:
        print("================== Thank you for using our Rent A Car ====================")
        break
    else:
        print("Enter the correct value")
```
